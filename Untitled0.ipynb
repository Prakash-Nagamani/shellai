```python
def filter_dp(dp):
  ans=list(list())
  ans.append(list(dp[0]))
  for i in range(1,len(dp)):
    if ans[-1][0]==dp[i][0]:
      ans[-1][1]+=dp[i][1]
    else:
      ans.append(list(dp[i]))
  ans=sorted(ans,key=lambda x:x[0])
  return ans
```

```python
def get_closest_supply_index(demand_index,distance):
  # now distance matrix is a global variable....make sure even demand_index is global
  distance =distance.to_numpy()
  return np.argsort(distance[demand_index])

def update_supply_remaining(ind,supply_remaining,supply,demand,supply_cap,supply_capacity):
    
    mini_energy=demand
    cost=[1.5,1]
    capacity=[400,200]
    mini_infra=get_minimum_infra(cost,capacity,mini_energy)
    cost=(1.5*mini_infra["FCS"]+mini_infra["SCS"])*600
    temp=mini_infra["FCS"]+mini_infra["SCS"]
    ut=0
    while supply_capacity[supply[ut]][0]<=supply_capacity[supply[ut]][1]+supply_capacity[supply[ut]][2]+temp:
      ut+=1
    ut=supply[ut]
    capacity_new=mini_infra["FCS"]*400+mini_infra["SCS"]*200
    # update supply_remaining
    supply_remaining.append(list([supply[ut],capacity_new]))
    supply_remaining=sorted(supply_remaining,key=lambda x:x[1])
    # update dict_suppy and dict_supply_rev
    dict_supply={}
    dict_supply_rev={}
    for i in range(len(supply_remaining)):
      dict_supply[i]=supply_remaining[i][1]
    for i in dict_supply:
      dict_supply_rev[dict_supply[i]]=i
    # update supply_cap
    supply_cap[supply[ut]]+=capacity_new
    # update supply_capacity
    supply_capacity[supply[ut]][0]+=temp
    supply_capacity[supply[ut]][1]+=mini_infra["FCS"]
    supply_capacity[supply[ut]][2]+=mini_infra["SCS"]

    return supply_remaining,supply_cap,supply_capacity,dict_supply,dict_supply_rev
   

def get_minimum_infra(cost,capacity,requirement):
  mini=min(capacity)
  if requirement<=mini:
      return {"FCS":0,"SCS":1}
  else:
      diff=requirement
      maxi=max(capacity)
      fcs=0
      scs=0
      while diff>maxi:
          diff-=maxi
          fcs+=1
      if diff<=mini:
          scs+=1
          return {"FCS":fcs,"SCS":scs}
      else:
          fcs+=1
          return {"FCS": fcs, "SCS": scs}

def get_infra_cost(demand_index,supply_remaining, unsatisfied_dp,distance, dict_supply_rev,dict_demand_rev):
  # now dict_supply_rev and dict_demand_rev is used to retrieve data from supply_remaining when you have global index....
  # and dict_supply and dict_demand is used to retrive global index with local index
  # step-1: find the demand
  demand=unsatisfied_dp[demand_index][1]
  # now, find the closest supply point
  supply=get_closest_supply_index(demand_index,distance)
  # now, find what is the minimum energy infrastructure has to be set up and its cost related to it...

  # now, we need to find the minimum number of fcs and scs we need to increase...such that the cost is minimum and requirement is satisfied
  
  
  # before that u need to understand that...there could be cases when the nearest supply point is not there in supply_remaining...
  # so if they are not present then we need to add them in supply_remaining....
  ut=[]
  for i in supply_remaining:
    ut.append(i[0])

  if supply[0] not in ut:
    supply_remaining,supply_cap,supply_capacity,dict_supply,dict_supply_rev=update_supply_remaining(supply[0],supply_remaining,supply,demand,supply_cap,supply_capacity)
    return mini_infra,cost,supply,supply_remaining,supply_cap,supply_capacity,dict_supply,dict_supply_rev
  else:
    mini_energy=abs(demand-supply_remaining[dict_supply_rev[supply[0]]][0])
    if mini_energy<=supply_remaining[dict_supply_rev[supply[0]]][0]:
      return {"FCS":0,"SCS":0},0,supply,supply_remaining,supply_cap,supply_capacity,dict_supply,dict_supply_rev
    else:
      cost=[1.5,1]
      capacity=[400,200]
      mini_infra=get_minimum_infra(cost,capacity,mini_energy)
      cost=(1.5*mini_infra["FCS"]+mini_infra["SCS"])*600
      return mini_infra,cost,supply,supply_remaining,supply_cap,supply_capacity,dict_supply,dict_supply_rev


def distribute_supply(ans,existing_infra,dict_supply,dict_demand,result_temp, 
                      unsatisfied_dp,cost_demand,supply_remaining,
                      supply_capacity,dict_supply_rev,dict_demand_rev,
                      distance):
  # here we have changed local index to global index in result_temp...
  # so if you want to used result_temp[0] and result_temp[1]...you can use it directly for global variables

  cost_dict={}
  for i in range(len(cost_demand)):
    cost_dict[i]=cost_demand[0]

  unsatisfied_dp_new=list(list())

  # note: unsatisfied_dp and cost_demand has same index and both of their indexes are global...so no need tp change the index..
  # note: supply_remaining is not global...it is local

  for i in range(len(unsatisfied_dp)):
    mini_infra,infra_cost,sorted_supply=get_infra_cost(i,supply_remaining, unsatisfied_dp,distance, dict_supply_rev,dict_demand_rev)
    distributing_cost=cost_demand[i][1]
    if infra_cost>distributing_cost or infra_cost==0:
      for j in result_temp:
        if supply_remaining[dict_supply_rev[j[1]]][0]>j[2] and j[2]>0:
          supply_remaining[dict_supply_rev[j[1]]][0]-=j[2]
          ans[j[0]][j[1]]+=j[2]
        elif supply_remaining[dict_supply_rev[j[1]]][0]<=j[2] and j[2]>0:
          rem=abs(supply_remaining[dict_supply_rev[j[1]]][0]-j[2])
          ans[j[0]][j[1]]+=supply_remaining[dict_supply_rev[j[1]]][0]
          supply_remaining[dict_supply_rev[j[1]]][0]=0
          unsatisfied_dp_new.append(list([j[0],rem]))
    else:
      # increase the infrastructure in supply_capacity and assign the demand to that supply point in ans matrix...
      total_inc_infra=mini_infra["FCS"]+mini_infra["SCS"]
      j=0
      while total_inc_infra>supply_cap[sorted_supply[j]]:
        j+=1
      if j!=0:
        j=sorted_supply[j]
        demand=unsatisfied_dp[i][1]
        # now, find the closest supply point
        supply=get_closest_supply_index(i,distance)
        # now, find what is the minimum energy infrastructure has to be set up and its cost related to it...
        print(len(supply_remaining),dict_supply_rev[supply[j]])
        mini_energy=abs(demand-supply_remaining[dict_supply_rev[supply[j]]][0])
        if mini_energy<=supply_remaining[dict_supply_rev[supply[j]]][0]:
          mini_infra={"FCS":0,"SCS":0}
          infra_cost=0
        else:
          # now, we need to find the minimum number of fcs and scs we need to increase...such that the cost is minimum and requirement is satisfied
          cost=[1.5,1]
          capacity=[400,200]
          mini_infra=get_minimum_infra(cost,capacity,mini_energy)
          infra_cost=(1.5*mini_infra["FCS"]+mini_infra["SCS"])*600
          
        
       
      elif j==0:
        j=sorted_supply[j] # making it global j
      
      # now increase the supply as per the values in mini_infra
      supply_cap[j][1]=mini_infra["SCS"]
      supply_cap[j][2]=mini_infra["FCS"]

      # update in ans......
      ans[dict_demand[i]][j]+=unsatisfied_dp[i]

      # now update supply remaining
      present=False
      for i1 in supply_remaining:
        if i1[1]==j:
          present=False
          break
      if present:
        supply_remaining[dict_supply_rev[j]]=mini_infra["FCS"]*400+mini_infra["SCS"]*200-unsatisfied_dp[i]
      else:
        supply_remaining.append(list([mini_infra["FCS"]*400+mini_infra["SCS"]*200-unsatisfied_dp[i],j]))
    
    supply_remaining_new=list(list())
    for i1 in supply_remaining:
      if i1[0]>0:
        supply_remaining_new.append(list([i1[0],i1[1]]))
    return ans,unsatisfied_dp_new,supply_remaining_new
```
